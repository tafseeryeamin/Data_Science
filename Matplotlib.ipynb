```python
#Draw a line in a diagram from position (0,0) to position (6,250):

x=np.array([0,6])
y=np.array([0,250])

plt.plot(x,y)
plt.show()
```

points = np.array([0, 6]): This creates a NumPy array representing the x-coordinates of the points to be plotted. In this case, the x-coordinates are 0 and 6.
ypoints = np.array([0, 250]): This creates a NumPy array for the y-coordinates of the points. The y-coordinates are 0 and 250.

**Plotting Without Line**

```python
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, 'o')# ading 'o' we can make plot without line
plt.show()
```

**Multiple Points**

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

Mark each point with a circle:

```python
plt.plot(xpoints,ypoints, marker="o")
plt.show()
```

```python
plt.plot(xpoints,ypoints, marker="*")
plt.show()
```

```python
plt.plot(ypoints, 'o:r')
plt.show()
```

```python
plt.plot(ypoints, 'o:g')
plt.show()
```

'r'	Red

```python
plt.plot(ypoints,'o:r', ms = 20)
plt.show()
```

# CREATING PIE CHART

```python
y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

#ADDING LABELS
plt.pie(y, labels = mylabels)
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```
