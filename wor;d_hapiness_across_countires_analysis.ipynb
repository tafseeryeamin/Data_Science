```python
df1 = pd.read_csv("/content/world-happiness-report-2024/World-happiness-report-updated_2024.csv", encoding='ISO-8859-1')
df1.head (15)
```

```python
selected_country = 'Afghanistan'
country_data = df1[df1['Country name'] == selected_country]


plt.figure(figsize=(10, 6))
plt.plot(country_data['year'], country_data['Life Ladder'], marker='o', color='b', linestyle='-', label='Life Ladder (Happiness Score)')
plt.title(f'Happiness Score Trend Over Years for {selected_country}', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Happiness Score (Life Ladder)', fontsize=12)
plt.grid(True)
plt.xticks(country_data['year'])
plt.legend()
plt.tight_layout()


plt.show()
```

It seems like happiness level was above 3.5 from 2008 to 2013. Later on it came down and bounced back in 2016.

```python

plt.figure(figsize=(10, 6))
plt.plot(country_data['year'], country_data['Log GDP per capita'], marker='o', color='g', linestyle='-', label='Log GDP per Capita')
plt.title(f'Log GDP per Capita Trend Over Years for {selected_country}', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Log GDP per Capita', fontsize=12)
plt.grid(True)
plt.xticks(country_data['year'])
plt.legend()
plt.tight_layout()

plt.show()
```

```python

plt.figure(figsize=(10, 6))
plt.plot(country_data['year'], country_data['Social support'], marker='o', color='r', linestyle='-', label='Social Support')
plt.title(f'Social Support Trend Over Years for {selected_country}', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Social Support', fontsize=12)
plt.grid(True)
plt.xticks(country_data['year'])
plt.legend()
plt.tight_layout()

plt.show()
```

It seems like social support was above 0.45 level till 2018 and later on it came down.

Conclution:We can say happiness level of afganistan came down from 2019 and it is visable that social support and gdp also came down. It was pretty parallal we can say. And heatmap also says that.
