```python
order_quantity=np.array(df["Order_Quantity"])
unit_cost=np.array(df[' Unit_Cost '])
unit_price=np.array(df[' Unit_Price '])

cost = np.multiply(order_quantity, unit_cost)
revenue = np.multiply(order_quantity, unit_price)
profit = revenue - cost

print("Cost:", cost)
print("Revenue:", revenue)
print("Profit:", profit)
```

```python
mean_profit = np.mean(profit)
std_profit = np.std(profit)

# Identify outliers
outliers = profit[(profit > mean_profit + 2 * std_profit) | (profit < mean_profit - 2 * std_profit)]
print("Outliers:", outliers)
```

```python
print("Min Profit:", np.min(profit))
print("Max Profit:", np.max(profit))
print("Median Profit:", np.median(profit))
print("Mean Profit:", np.mean(profit))
print("Std Deviation:", np.std(profit))
```

```python
cumulative_profit = np.cumsum(profit)
print("Cumulative Profit (first 10 values):", cumulative_profit[:10])
```

```python
# Profit margin = (Profit / Revenue) * 100
profit_margin = np.divide(profit, revenue) * 100
print("Profit Margin (%):", np.around(profit_margin, 2))
```

```python
# Reshape profits into 4 weeks (approximate)
weekly_profits = profit[:28].reshape(4, 7)  # First 28 entries as 4 weeks
print("Weekly Profits:\n", weekly_profits)
```
