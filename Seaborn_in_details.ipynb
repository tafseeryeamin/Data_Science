```python
#Import Seaborn

import seaborn as sns

#Apply theme

sns.set_theme()

#load an example data
tips=sns.load_dataset("tips")
tips
```

A **relational plot** using the sns.relplot() function from the seaborn library, which is a popular Python visualization library based on Matplotlib

```python
#Create a visualization

sns.relplot(
    data=tips,              # Data source is the 'tips' dataset
    x="total_bill",         # X-axis is the 'total_bill' column, representing the total bill amount
    y="tip",                # Y-axis is the 'tip' column, representing the tip amount
    col="time",             # Create separate subplots for each unique value in the 'time' column (Lunch, Dinner)
    hue="smoker",           # Use different colors for smokers and non-smokers based on the 'smoker' column
    style="smoker",         # Different marker styles (shapes) for smokers and non-smokers
    size="size",            # Vary the size of markers based on the 'size' column (group size at the table)
)
```

**A high-level API for statistical graphics**

```python
dots= sns.load_dataset("dots")
dots
```

The relplot() function has a convenient kind parameter that lets you easily switch to this alternate representation:

```python
sns.relplot(
    data=dots,
    kind="line",
    x="time",
    y="firing_rate",
    col="align",
    hue="choice",
    style="choice",
    size="coherence",
    facet_kws=dict(sharex=False),
)
```

# **Statistical estimation**

it is possible to enhance a scatterplot by including a linear regression model (and its uncertainty) using **lmplot():**

```python
sns.lmplot(data=tips, x="total_bill", y="tip", col="time", hue="smoker")
```

Many seaborn functions will automatically perform the statistical estimation that is necessary to answer these questions:

```python
f=sns.load_dataset("fmri")

f
```

```python
sns.relplot(
    data=f,
    kind="line",
    x="timepoint",
    y="signal",
    col="event",
    hue="region",
    )
```

# Distributional Respresentaations

The seaborn function displot() supports several approaches to visualizing distributions.

```python
sns.displot(
    data=tips,
    x="total_bill",
    col="time",
    kde=True,   #Kernel Density Estimate-used to estimate the probability density function (PDF) of a continuous random variable
)
```

# Plots for categorical data


It can be accessed through catplot()

```python
sns.catplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="smoker",
    kind="bar", #swarm,bar,violin
)
```

# Multivariate views on complex datasets


Can be accessed usiing jointplot()

```python
penguins=sns.load_dataset("penguins")


penguins
```

```python
sns.jointplot(
    data=penguins,
    x="flipper_length_mm",
    y="bill_length_mm",
    hue="species",
)
```

```python
sns.pairplot(
    data=penguins,
    hue="species"

)
```
